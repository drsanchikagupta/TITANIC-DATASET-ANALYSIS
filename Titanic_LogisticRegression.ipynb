```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LogisticRegression
```

```python
X = pd.read_csv('titanic_train.csv')
Xtest=pd.read_csv('titanic_test.csv')
```

```python
X.head()
```

```python
Y = X.Survived
```

```python
Y.head()
```

```python
X['Age'].describe()
```

```python
def Age_to_Val(Passenger):
    age= Passenger
    
    if 0<=age<1: 
        return 1
    elif 1<=age<6:
        return 2
    elif 6<=age<15:
        return 3
    elif 15<=age<30:
        return 4
    elif 30<=age<60:
        return 5
    else:
        return 6
```

```python
X['Person']= X['Age'].apply(Age_to_Val)
```

```python
Xtest['Person']= Xtest['Age'].apply(Age_to_Val)
Xtest.head()
```

```python
X.head()
```

```python
def get_dummies(df, col_name):
    dummies=pd.get_dummies(df[col_name], prefix=col_name)
    df=pd.concat([dummies, df], axis=1)
    return df
```

```python
X= get_dummies(X, 'Person')
X= get_dummies(X,'Pclass')
```

```python
Xtest= get_dummies(Xtest, 'Person')
Xtest= get_dummies(Xtest,'Pclass')
```

```python
X.loc[X.Sex=='male', 'Sex']=1
X.loc[X.Sex=='female', 'Sex']=0

X= get_dummies(X,'Sex')
X.head()
```

```python
Xtest.loc[Xtest.Sex=='male', 'Sex']=1
Xtest.loc[Xtest.Sex=='female', 'Sex']=0

Xtest= get_dummies(Xtest,'Sex')
Xtest.head()
```

```python
X['Alone']= X['SibSp']+X['Parch']

X.loc[X.Alone>0, 'Alone']=1
X.loc[X.Alone==0, 'Alone']=0

X= get_dummies(X,'Alone')
X.head()
```

```python
Xtest['Alone']= Xtest['SibSp']+Xtest['Parch']

Xtest.loc[Xtest.Alone>0, 'Alone']=1
Xtest.loc[Xtest.Alone==0, 'Alone']=0

Xtest= get_dummies(Xtest,'Alone')
Xtest.head()
```

```python
X_train=X.drop(['Name','Age','Ticket','Fare','Cabin','Person','Survived','Pclass','Sex','SibSp', 'Parch','Alone','Embarked'], axis=1)
X_train.head()
```

```python
X_test=Xtest.drop(['Name','Age','Ticket','Fare','Cabin','Person','Pclass','Sex','SibSp', 'Parch','Alone','Embarked'], axis=1)
X_test.head()
```

```python
lreg=LogisticRegression(solver='lbfgs', max_iter=2500)
```

```python
lreg.fit(X_train,Y)
```

```python
x_train, x_test, y_train, y_test= sklearn.model_selection.train_test_split(X_train, Y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
lreg.fit(x_train, y_train)
```

```python
lreg.score(x_train, y_train)
```

```python
pred= lreg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, pred)
pd.DataFrame(conf_matrix, columns=['Survived', 'Died'], index=[['Survived', 'Died']])
```

```python
#Using Cross Validation for More Accurate Error Measurement

from sklearn.model_selection import cross_val_score

scores = cross_val_score(lreg, x_train, y_train, cv=20)
np.mean(scores)
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y)
holdout_predictions = lr.predict(X_test)
holdout_predictions
```

```python
holdout_predictions.shape
```

```python
holdout_ids = X_test["PassengerId"]
submission_df = {"PassengerId": holdout_ids,
                 "Survived": holdout_predictions}
submission = pd.DataFrame(submission_df)
```

```python
submission
```

```python
submission.to_csv('titanic_submission.csv', index=False)
```
